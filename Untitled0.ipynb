```python
import seaborn as sns
sns.pairplot(df,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',size=10,aspect=0.5)
x=df.iloc[:,[1,2]].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
import statsmodels.api as sm
x_train_sm=sm.add_constant(x_train)
model=sm.OLS(y_train,x_train_sm).fit()
print("model coefficient=",model.params)
print("residual standards error=",model.rsquared)
print("F-statistics=",model.fvalue)
print("T-statistics=",model.tvalues)
print("residual sum of squares=",model.ssr)

```
